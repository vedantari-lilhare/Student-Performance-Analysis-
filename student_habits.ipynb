```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize']=(12,6)
```

```python
numerical_cols =['age','study_hours_per_day','social_media_hours','netflix_hours','attendance_percentage','sleep_hours','exercise_frequency','internet_quality','mental_health_rating','exam_score']
for col in numerical_cols:
    sns.histplot(student_habits_update_dataset[col],kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_cols = ['gender','part_time_job','parental_education_level','internet_quality','extracurricular_participation']
for col in categorical_cols:
    sns.countplot(x=student_habits_update_dataset[col])
    plt.title(f'count of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
for col in categorical_cols:
    sns.boxplot(x=student_habits_update_dataset[col], y=student_habits_update_dataset['exam_score'])
    plt.title(f'Exam score by {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
scatter_features =['study_hours_per_day','social_media_hours','sleep_hours']

for col in scatter_features:
    sns.scatterplot(data=student_habits_update_dataset, x=col, y='exam_score', hue='gender')
    plt.title(f'{col} vs Exam Score')
    plt.show()
```

```python
for col in student_habits_update_dataset[numerical_cols]:
    if student_habits_update_dataset[col].dtype =='object':
       student_habits_update_dataset[col] = student_habits_update_dataset[col].astype('category').cat.codes
        
corr = student_habits_update_dataset[numerical_cols].corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot = True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
selected_features =['study_hours_per_day','sleep_hours','social_media_hours','mental_health_rating','exam_score']
sns.pairplot(student_habits_update_dataset[selected_features], diag_kind='kde')
plt.show()
```

```python
grouped = student_habits_update_dataset.groupby(['parental_education_level','gender'])['exam_score'].mean().reset_index()
sns.barplot(data=grouped, x='parental_education_level', y='exam_score', hue='gender')
plt.title('Mean Exam Score by Parental Education and Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.violinplot(data=student_habits_update_dataset, x='diet_quality', y='exam_score')
plt.title('Exam Score by Diet Quality')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

cluster_data = student_habits_update_dataset[numerical_cols].dropna()
scaler = StandardScaler()
scaled = scaler.fit_transform(cluster_data)

kmeans = KMeans(n_clusters=3, random_state=42)
cluster_data['cluster'] = kmeans.fit_predict(scaled)

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled)
cluster_data['PCA1'] =pca_result[:,0]
cluster_data['PCA2'] =pca_result[:,1]

sns.scatterplot(data=cluster_data, x='PCA1', y='PCA2', hue='cluster', palette='Set2')
plt.title("Cluster of Students (PCA + KMeans)")
plt.show()
```
